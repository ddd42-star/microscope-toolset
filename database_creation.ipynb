##### Now contextualize all the differents function from the pymmcore api documentation to build different reference between shared functions

```python
from usage.contextualization_pymmcore import ContextualizationOutput

def contextualize_chunks(document, chunk):

    api_key='<write here your api key>'

    client = OpenAI(api_key=api_key)



    PROMPT = """
    Whole document:
    {doc_content}

    Here is the chunk we want to situate within the whole document:
    {chunk_content}
    """
    user_query = """
    Please give a short succinct context to situate this chunk within the overall document context for the purposes of improving search retrieval of the chunk.
    Answer only with the succinct context and nothing else.
    """
    prompt = PROMPT.format(doc_content=document, chunk_content=chunk)

    response = client.responses.parse(
        model="gpt-4.1-mini",
        input=[{"role": "system", "content": prompt}, {"role": "user", "content": user_query}],
        text_format=ContextualizationOutput
    )

    #parsed_response = parse_output(response.choices[0].message.content)

    parsed_response = response.output_text

    # The output should remain a json object that we will save inside the databases

    return parsed_response
```

```python
from tqdm import tqdm

new_list = []

for content in changed_list:
    for inside in content:
        new_list.append(json.dumps(inside))
# join all together
all_doc = '\n'.join(new_list)

# load docs
list_of_all_document = sorted(Path(api_path).glob("*_updated.json"), key=lambda x: x.stem)

print(list_of_all_document)

for doc in tqdm(list_of_all_document):

    print(f"Processing {doc}")

    with open(doc, "r", encoding="utf-8") as f:
        text_json = f.read()
    f.close()

    tmp_list = json.loads(text_json)

    new_doc_list = []
    # iterate through each chunk
    for chunk in tqdm(tmp_list):
        contextualize_text = contextualize_chunks(all_doc, json.dumps(chunk))
        #print(contextualize_text)

        chunk["contextualize_text"] = json.loads(contextualize_text)["message"]

        new_doc_list.append(chunk)

    # write the final file
    with open(f"{api_path}/{doc.name.split('.')[0]}_finale.json", "w", encoding="utf-8") as f:
        f.write(json.dumps(new_doc_list))
    f.close()

    print(f"finish {doc}")
```

##### Creation of the database of all publication available before 24.08.2025

We extracted the text from all the publication of the Pertz lab available online. Then running the file *create_database_from_publications*, we created first single images for each page of each publication, then we extracted the text from this page, filtering out authors, pages number, publication journal. Afterwords we created a single markdown file with all the text. The intrinsic structure of a publication allowed us to maintain the differents main parts like 'introduction', 'materials and methods', 'results' and 'conclusion'. We choose to not maintain any figure and any figure description because it was just adding text in multiple cases in the middle of other sections, making harder to maintain a simple structure. The bigger danger was that later, after chunking the text, it would had caused hallucination.

##### Creation of the different indexes in our server

After preparing the different text, we created three different indices in our database.

```python
pdf_publication = "publication-collection"
micromanager_collection = "devices-collection"
api_collection = "documentation-microscope"
```

```python
from pathlib import Path
from semantic_text_splitter import MarkdownSplitter
from databases.elasticsearch_db import ElasticSearchDB
from openai import OpenAI
from tqdm import tqdm
```

```python
es_client = ElasticSearchDB()
```

```python
device_files = "./micro_manager"

markdown_list = sorted(Path(device_files).glob("*.md"), key=lambda x: x.stem)

api_key='<write her your api key>'

client = OpenAI(api_key=api_key)

# open file, read content, create dict
splitter = MarkdownSplitter(5000)

content_to_digest = []

for file in tqdm(markdown_list):
    print(f"Processing {file}")

    file_name = file.name.split('.')[0]

    with open(file, 'r', encoding='utf-8') as f:
        markdown_text = f.read()
    f.close()

    # create markdown chunks
    chuncks_list = splitter.chunks(markdown_text)

    for chunk_index, chunk_content in enumerate(chuncks_list):

        # generate embedding
        print(f"Generating embedding chunk {chunk_index}")
        response = client.embeddings.create(input=chunk_content, model="text-embedding-3-small", dimensions=512)
        print(f"Done with chunk {chunk_index}!")
        embedding = response.data[0].embedding

        doc_dict = {
            "doc_id": f"{file_name}_chunk_{chunk_index}",
            "content": chunk_content,
            "chunk_id": f"{chunk_index}",
            "filename": f"{file_name}",
            "embedding": embedding,
        }
        # add object to the list
        content_to_digest.append(doc_dict)


```

```python
content_to_digest[0]
```

```python
# create the micromanage device collection
es_client.create_default_index(index_name=micromanager_collection, doc_id={"type":"keyword"}, content={"type":"text"}, chunk_id={"type":"keyword"}, filename={"type":"keyword"})
```

```python
# indexes the micromanage collection
es_client.index_documents(index_name=micromanager_collection, documents=content_to_digest)
```
